## **MatPlotLib**

Matplotlib is a plotting library for the Python programming language and its numerical mathematics extension NumPy. It provides an object-oriented API for embedding plots into applications using general-purpose GUI toolkits like Tkinter, wxPython, Qt, or GTK. There is also a procedural "pylab" interface based on a state machine (like OpenGL), designed to closely resemble that of MATLAB, though its use is discouraged. SciPy makes use of Matplotlib.

Matplotlib was originally written by John D. Hunter.

Matplotlib supports various types of 2 dimensional and 3 dimensional plots. The support for two dimensional plots is robust. The support for three dimensional plots was added later and while it is good, it is not as robust as 2 dimensional plots.

```python
from scipy.stats import norm
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

import warnings
warnings.filterwarnings("ignore")
```

```python
x = np.arange(-3, 3, 0.1)
y = norm.pdf(x)
df = pd.DataFrame({'x': x, 'y': y})
df.head()
```

```python
x= df['x']
y= df['y']
```

```python
plt.scatter(x,y) #Scatter Plot
plt.title("Normal Distribution")
plt.show() # Displaying Plot
```

```python
plt.plot(x,y) #Line Plot
plt.show()
```

### **Box & Whisker Plot**

Useful for visualizing the spread & skew of data.

The red line represents the median of the data, and the box represents the bounds of the 1st and 3rd quartiles.

So, half of the data exists within the box.

The dotted-line "whiskers" indicate the range of the data - except for outliers, which are plotted outside the whiskers. Outliers are 1.5X or more the interquartile range.

This example below creates uniformly distributed random numbers between -40 and 60, plus a few outliers above 100 and below -100:

<pre><code>     Q1-1.5IQR   Q1   median  Q3   Q3+1.5IQR
                  |-----:-----|
  o      |--------|     :     |--------|    o  o
                  |-----:-----|
flier             <----------->            fliers
                       IQR</code></pre>

```python
import pandas as pd
uniformSkewed = np.random.rand(100) * 100 - 40
high_outliers = np.random.rand(10) * 50 + 100
low_outliers = np.random.rand(10) * -50 - 100
data = np.concatenate((uniformSkewed, high_outliers, low_outliers))
df = pd.DataFrame({'data': data}, columns=['data'])
print(df)
df.describe()
```

```python
plt.boxplot(df['data'],labels=df.columns, vert=True, showfliers=True, meanline=True)
plt.show()
```

### **3D plot in Matplotlib**

```python
from mpl_toolkits.mplot3d import Axes3D
```

```python
# Generate random data
x = np.random.rand(50)
y = np.random.rand(50)
z = np.random.rand(50)
```

```python
fig = plt.figure(figsize=(8,6))  # Create a figure with a white background
ax = fig.add_subplot(projection='3d')  # Create a 3D axis
ax.scatter(x, y, z, c='r', marker='o')

ax.set_xlabel("X Axis")
ax.set_ylabel("Y Axis")
ax.set_zlabel("Z Axis")
ax.set_title("3D Line Plot")

plt.title("3D Scatter Plot")
plt.tight_layout()

plt.show()
```

### **Subplots**

```python
x=np.arange(-3,3,0.1)
data = np.random.randn(1000)

plt.figure(figsize=(8,6)) 

plt.subplot(221)  # This is to create a 2x2 plot 1st position
plt.plot(x,x**2, color='blue', linestyle='--')
plt.xlabel('Event1')
plt.ylabel('Event2')
plt.title('Fun A')

plt.subplot(2, 2, 2)
plt.scatter(x, norm.pdf(x), color='yellow', marker='*')
plt.xlabel('Event1')
plt.ylabel('Event2')
plt.title('Fun B')

plt.subplot(2, 2, 3)
plt.hist(data, bins=10, color='skyblue', edgecolor='black', alpha=0.7)
plt.xlabel('Event1')
plt.ylabel('Event2')
plt.title('Fun C')

plt.subplot(2, 2, 4)
plt.boxplot(data, vert=False, patch_artist=True)
plt.xlabel('Event1')
plt.title('Fun D')

plt.tight_layout() # Adjust the plot to fit the figure area

plt.show()
```
