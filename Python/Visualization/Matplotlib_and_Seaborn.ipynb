### **Pie Chart**

```python
values =[12,30, 4, 32, 14]

labels = ['A', 'B', 'C', 'D', 'E']

colors = ['red', 'blue', 'green', 'cyan', 'm']

explode = [0,0.1,0,0.2,0]

plt.pie(values, labels=labels,autopct='%1.f%%', radius=0.7, colors=colors, explode=explode)

plt.legend(title='Label')

plt.title('Pie Chart')

plt.show()
```

### **Bar Plot**


```python
x = ['A', 'B', 'C', 'D', 'E']
y = [10,20,30,40,25]

plot = plt.bar(x,y, color=['r', 'g', 'b', 'm', 'y'], edgecolor='k')
plt.bar_label(plot, label_type='center', rotation=90, fmt='%.1f')

plt.plot(x,y, marker='*', color='r')

plt.xlabel('X Label')
plt.ylabel('Y Label')

plt.grid(axis='both')

plt.legend(plot, x)

plt.title("Vertical Bar chart")

plt.show()
```

```python
x = ['A', 'B', 'C', 'D', 'E']
y = [10,20,30,40,25]

plot = plt.barh(x,y, color=['r', 'g', 'b', 'm', 'y'], edgecolor='k')
plt.bar_label(plot, label_type='center', rotation=90, fmt='%.1f')

plt.xlabel('Y Label')
plt.ylabel('X Label')

plt.grid(axis='both')

plt.legend(plot, x)

plt.title("Horizontal Bar chart")

plt.show()
```

### **box or whisker plot**

<pre><code>     Q1-1.5IQR   Q1   median  Q3   Q3+1.5IQR
                  |-----:-----|
  o      |--------|     :     |--------|    o  o
                  |-----:-----|
flier             <----------->            fliers
                       IQR</code></pre>

```python
import pandas as pd
uniformSkewed = np.random.rand(100) * 100 - 40
high_outliers = np.random.rand(10) * 50 + 100
low_outliers = np.random.rand(10) * -50 - 100

data = np.concatenate((uniformSkewed, high_outliers, low_outliers))

df = pd.DataFrame({'data': data})
df.head()
```

```python
df.describe()
```

```python
plt.boxplot(df['data'], vert=False, showfliers=True, meanline=True, labels=df.columns)
plt.show()
```

### **Subplot**

```python
x = np.arange(-3,3,0.1)
data = np.random.randn(1000)

plt.figure(figsize=(8,6))

plt.subplot(2,2,1)
plt.plot(x, norm.pdf(x), c='b', linestyle='--')
plt.xticks([-3,0,1,-4])
plt.title('Line Plot')

plt.subplot(2,2,2)
plt.scatter(x, norm.pdf(x, 1, 0.5), c='y', marker='*')
plt.grid(axis='both')
plt.title('Scatter plot')
plt.title("Histogram")

plt.subplot(223)
plt.hist(data, bins=10, color='g', edgecolor='k', alpha=0.7)
plt.xlabel('Data')
plt.ylabel('Frequency')

plt.subplot(224)
plt.boxplot(data, vert=False, patch_artist=True)
plt.xlabel('Event1')
plt.title('Boxplot')

plt.tight_layout()

plt.show()
```
